```python
import pandas as pd
import numpy as np
url='https://drive.google.com/file/d/1JMYqXipZpz9Y5-vyxvLEO2Y1sRBxqu-U/view?usp=sharing'
file_id=url.split('/')[-2]
dwn_url='https://drive.google.com/uc?id=' + file_id
df = pd.read_csv(dwn_url)
```

```python
df = df.drop(['Специализация', 'Университет', 'Предметная.область', 'Валюта', 'cls'], axis=1)
df
```

```python
df = df.dropna()
```

```python
df.shape
```

```python
new_df = df[df['Язык.программирования'] == 'Python']
new_df
```

```python
new_df.shape
```

```python
group_df = new_df.groupby('Должность').min()
group_df
```

```python
agg_df = df.groupby('Должность').agg({'Зарплата.в.месяц': ['min', 'max']})
agg_df
```

```python
def fill_avg_salary(df):
    avg_salary = df['Зарплата.в.месяц'].mean()
    return avg_salary

agg_df['avg'] = agg_df.apply(fill_avg_salary, axis=1)

agg_df
```

```python
result = agg_df['avg'].describe()
result
```

```python
result.to_csv("avg_salary.csv")
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.plot(result)
```
